```python
from collections import Counter
import re

# Function to extract keywords from free-text columns
def extract_keywords(text_series, top_n=10):
    text = " ".join(text_series.dropna()).lower()  # Convert to lowercase & remove NaN
    words = re.findall(r'\b[a-zA-Z]{3,}\b', text)  # Extract words (ignore numbers/special chars)
    common_words = Counter(words).most_common(top_n)  # Get most common words
    return dict(common_words)

# Process each free-text column
correction_tags = extract_keywords(df1["CORRECTION_VERBATIM"], top_n=10)
customer_tags = extract_keywords(df1["CUSTOMER_VERBATIM"], top_n=10)
casual_tags = extract_keywords(df2["CAUSAL_VERBATIM"], top_n=10)




# Display extracted keywords
print("Correction Verbatim Keywords:", correction_tags)
print("Customer Verbatim Keywords:", customer_tags)
print("Causal Verbatim Keywords:", casual_tags)


```

```python
def extract_sentences(column, top_n=10):
    text_data = df1[column].dropna().str.lower()  # Convert to lowercase and remove NaN values
    sentences = list(text_data)
    
    # Tokenizing words and getting the most common ones
    words = []
    for text in text_data:
        words.extend(re.findall(r'\b\w+\b', text))  # Extract words using regex
    common_words = [word for word, _ in Counter(words).most_common(top_n)]
    
    # Extract sentences that contain the most common words
    extracted_sentences = [sentence for sentence in sentences if any(word in sentence for word in common_words)]
    
    return extracted_sentences

# Extract full sentences from CUSTOMER_VERBATIM and CAUSAL_VERBATIM
customer_sentences = extract_sentences("CUSTOMER_VERBATIM", top_n=10)
#causal_sentences = extract_sentences("CAUSAL_VERBATIM", top_n=10)

# Display results
print("Extracted Sentences from CUSTOMER_VERBATIM:")
for sentence in customer_sentences[:10]:  # Display top 10 sentences
    print("-", sentence)

# print("\nExtracted Sentences from CAUSAL_VERBATIM:")
# for sentence in causal_sentences[:10]:  # Display top 10 sentences
#     print("-", sentence)
```
