```python

from googletrans import Translator

def translate_column(df, column_name, target_lang="en"):
    """
    Translate an entire column in a DataFrame using Google Translate.
    
    Parameters:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The column to be translated.
        target_lang (str): The target language (default is English 'en').

    Returns:
        pd.DataFrame: The DataFrame with an additional translated column.
    """
    translator = Translator()
    
    def translate_text(text):
        """Helper function to translate a single text entry."""
        try:
            return translator.translate(text, dest=target_lang).text
        except Exception as e:
            print(f"Error translating: {text}, Error: {e}")
            return text  # Return the original text if translation fails

    # Apply translation to the column
    work_df['Correction' + "_translated"] = df["Correction"].astype(str).apply(translate_text)
    
    return df

# Example usage:

```

```python
#seting the white grid background
sns.set_style("whitegrid")
```

```python
# Plot histograms for numerical columns
merged_df.hist(figsize=(12, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```python
# Boxplot to detect outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_df.select_dtypes(include=['float64', 'int64']), palette="coolwarm")
plt.title("Boxplot of Numerical Features")
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(merged_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Top 10 complaints
top_complaints = merged_df['Complaint'].value_counts().head(10)
top_complaints

```

```python
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'])

# Aggregate repair count per day
orders_per_day = merged_df.groupby(merged_df['Order Date'].dt.date).size()

plt.figure(figsize=(12, 6))
orders_per_day.plot(marker="o", linestyle="-", color="green")

plt.title("Repair Orders Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()


```
